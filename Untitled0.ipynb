# Task
Calculate the recency for each customer by finding their last purchase date relative to a reference date of '2011-12-10', and then calculate the recency in days.

## Calcular Recencia

### Subtask:
Calcula la última fecha de compra para cada cliente, define la fecha de referencia como '2011-12-10', y luego calcula la recencia en días. Este paso convertirá 'InvoiceDate' a formato de fecha para realizar los cálculos correctamente.


**Reasoning**:
First, I'll ensure the 'InvoiceDate' column is in datetime format to allow for date calculations. Then, I'll calculate the latest purchase date for each customer. Finally, I will define the reference date and compute the recency for each customer by subtracting their last purchase date from the reference date.



```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

last_purchase_date = df.groupby('CustomerID')['InvoiceDate'].max()

reference_date = pd.to_datetime('2011-12-10')

recency = (reference_date - last_purchase_date).dt.days

print("Últimas fechas de compra por cliente (head):")
print(last_purchase_date.head())
print("\nRecencia calculada para cada cliente (head):")
print(recency.head())
```

**Reasoning**:
The previous step calculated the recency for each customer. Now, I will integrate this 'recency' data into the main DataFrame `df` by mapping the calculated recency values to their respective customers, creating a new column named 'Recency'.



```python
df = df.merge(recency.rename('Recency'), on='CustomerID', how='left')

print("DataFrame with new 'Recency' column (head):")
display(df.head())
```

## Calcular Frecuencia y Valor Monetario

### Subtask:
Agrupa el DataFrame por 'CustomerID' para calcular la frecuencia (conteo de transacciones únicas) y el valor monetario (suma de 'Revenue Total') para cada cliente.


**Reasoning**:
I will group the DataFrame by 'CustomerID' to calculate the frequency of unique 'InvoiceNo' and the sum of 'Revenue Total' for each customer, storing the results in a new DataFrame named `rfm_f`.



```python
rfm_f = df.groupby('CustomerID').agg(
    Frequency=('InvoiceNo', 'nunique'),
    MonetaryValue=('Revenue Total', 'sum')
).reset_index()

print("Primeras filas del DataFrame 'rfm_f' con Frecuencia y Valor Monetario:")
display(rfm_f.head())
```

**Reasoning**:
To complete the RFM analysis, I will merge the calculated 'recency' data (which is a Series with 'CustomerID' as index) into the `rfm_f` DataFrame, which already contains 'Frequency' and 'MonetaryValue'. This will create a consolidated RFM DataFrame for each customer.



```python
rfm_f = rfm_f.merge(recency.rename('Recency'), on='CustomerID', how='left')

print("DataFrame RFM final (head):")
display(rfm_f.head())
```

## Asignar Puntuaciones RFM

### Subtask:
Asigna puntuaciones del 1 al 5 a las métricas de Recencia, Frecuencia y Valor Monetario utilizando cuantiles. Se invertirá la puntuación de Recencia para que un valor más bajo (más reciente) obtenga una puntuación más alta. Se mostrarán las primeras filas del `rfm_df` con las nuevas puntuaciones.


```python
import numpy as np

# For Recency, lower values are better (higher score), so we invert the percentile rank
rfm_f['R_Score'] = rfm_f['Recency'].rank(method='first', pct=True).apply(lambda x: int(np.ceil((1 - x) * 5)))

# For Frequency, higher values are better (higher score)
rfm_f['F_Score'] = rfm_f['Frequency'].rank(method='first', pct=True).apply(lambda x: int(np.ceil(x * 5)))

# For MonetaryValue, higher values are better (higher score)
rfm_f['M_Score'] = rfm_f['MonetaryValue'].rank(method='first', pct=True).apply(lambda x: int(np.ceil(x * 5)))

print("DataFrame RFM con puntuaciones (head):")
display(rfm_f.head())
```

## Calcular Puntuación RFM Total y Segmentos

### Subtask:
Calcula una puntuación RFM combinada sumando las puntuaciones individuales de R, F y M. Luego, define una función para asignar segmentos de clientes basados en estas puntuaciones combinadas y aplica esta función para crear una nueva columna 'RFM_Segment'.


**Reasoning**:
I will calculate the total RFM score by summing the individual R, F, and M scores, define a function to categorize customers into segments based on this total score, and then apply this function to create a new 'RFM_Segment' column. Finally, I will display the head of the DataFrame to show the new columns.



```python
rfm_f['RFM_Score'] = rfm_f['R_Score'] + rfm_f['F_Score'] + rfm_f['M_Score']

def assign_rfm_segment(score):
    if score >= 13:
        return 'Champions'
    elif score >= 10:
        return 'Loyal Customers'
    elif score >= 7:
        return 'Potential Loyalists'
    else:
        return 'At Risk'

rfm_f['RFM_Segment'] = rfm_f['RFM_Score'].apply(assign_rfm_segment)

print("DataFrame RFM con puntuación total y segmentos (head):")
display(rfm_f.head())
```

```python
from datetime import date

# Definir las fechas de inicio y fin para la ventana de observación
observation_start_date = date(2010, 12, 1)
observation_end_date = date(2011, 9, 30)

# Definir las fechas de inicio y fin para la ventana de predicción
prediction_start_date = date(2011, 10, 1)
prediction_end_date = date(2011, 12, 9)

# Calcular la duración de la ventana de observación en meses
num_months_observation = (observation_end_date.year - observation_start_date.year) * 12 + \
                         (observation_end_date.month - observation_start_date.month)

print(f"Ventana de Observación: {observation_start_date} a {observation_end_date}")
print(f"Ventana de Predicción: {prediction_start_date} a {prediction_end_date}")
print(f"Duración de la Ventana de Observación: {num_months_observation} meses")
```

```python
split_date = pd.to_datetime('2011-09-01')

train_data = df[df['InvoiceDate'] < split_date]
test_data = df[df['InvoiceDate'] >= split_date]

print(f"Forma de train_data (ventana de observación): {train_data.shape}")
print(f"Forma de test_data (ventana de predicción): {test_data.shape}")
```

```python
rfm_clv_features = train_data.groupby('CustomerID').agg(
    Recency=('InvoiceDate', lambda x: (split_date - x.max()).days),  # Días desde la última compra en train_data
    Frequency=('InvoiceNo', 'nunique'),  # Número de compras únicas
    Monetary=('Revenue Total', 'sum'),  # Gasto total
    Primera_Compra=('InvoiceDate', 'min') # Primera compra para antigüedad
).reset_index()

# Calcular antigüedad del cliente
rfm_clv_features['Antiguedad'] = (split_date - rfm_clv_features['Primera_Compra']).dt.days

# Calcular Ticket Promedio
rfm_clv_features['Ticket promedio'] = rfm_clv_features['Monetary'] / rfm_clv_features['Frequency']

# Calcular Frecuencia Media
# num_months_observation was defined previously and used for the overall observation window duration
rfm_clv_features['Frecuencia media'] = rfm_clv_features['Frequency'] / num_months_observation

# Eliminar la columna 'Primera_Compra' ya que ya se usó para 'Antiguedad'
rfm_clv_features = rfm_clv_features.drop(columns=['Primera_Compra'])

print("Primeras filas del DataFrame con RFM y variables adicionales:")
display(rfm_clv_features.head())
```

## 1. Implementación del modelo BG/NBD

### Subtask: Preparar datos en formato BG/NBD

El primer paso para implementar el modelo BG/NBD es transformar los datos transaccionales de `train_data` al formato de resumen requerido por la librería `lifetimes`. Este formato incluye la frecuencia (número de compras repetidas), recencia (tiempo entre la primera y la última compra), y `T` (tiempo desde la primera compra hasta el final del período de observación).

```python
from lifetimes.utils import summary_data_from_transaction_data

# 'Revenue Total' fue la columna calculada para el valor monetario
# 'observation_end_date' fue definida previamente como la fecha de fin de la ventana de observación
summary_bgnbd = summary_data_from_transaction_data(
    train_data,
    customer_id_col='CustomerID',
    datetime_col='InvoiceDate',
    monetary_value_col='Revenue Total', # Usamos 'Revenue Total' como valor monetario
    observation_period_end=observation_end_date # Usamos la fecha de fin de observación
)

print("Primeras filas del DataFrame de resumen BG/NBD:")
display(summary_bgnbd.head())
```

### Subtask: Entrenar el modelo BG/NBD

Con los datos preparados en el formato `summary_bgnbd`, el siguiente paso es entrenar el modelo Beta-Geometric/Negative Binomial Distribution (BG/NBD). Este modelo aprenderá los patrones de compra y abandono de los clientes a partir de la frecuencia, recencia y la antigüedad `T`.

```python
from lifetimes import BetaGeoFitter

bgf = BetaGeoFitter(penalizer_coef=0.01)
bgf.fit(summary_bgnbd['frequency'], summary_bgnbd['recency'], summary_bgnbd['T'])

print("Resumen del modelo BG/NBD entrenado:")
print(bgf.summary)
```

### Subtask: Predecir compras futuras

Una vez que el modelo BG/NBD está entrenado, podemos usarlo para predecir el número de compras que cada cliente realizará en un período futuro. Aquí usaremos un período de 120 días (aproximadamente 4 meses) para la predicción, como se sugirió.

```python
# t es el número de unidades de tiempo futuras para predecir.
# En este caso, si T es en días, t=120 es para 120 días futuros.
summary_bgnbd['predicted_purchases'] = bgf.conditional_expected_number_of_purchases_up_to_time(
    t=120,  # 120 días futuros (aproximadamente 4 meses)
    frequency=summary_bgnbd['frequency'],
    recency=summary_bgnbd['recency'],
    T=summary_bgnbd['T']
)

print("Primeras filas del DataFrame con compras futuras predichas:")
display(summary_bgnbd.head())
```

```python
summary_gg = summary_bgnbd[(summary_bgnbd['frequency'] > 0) & (summary_bgnbd['monetary_value'] > 0)]

print("Primeras filas del DataFrame filtrado para Gamma-Gamma:")
display(summary_gg.head())
```

## 2. Implementación del modelo Gamma–Gamma

### Subtask: Filtrar datos para el modelo Gamma–Gamma

Para el modelo Gamma–Gamma, es importante considerar solo a los clientes que han realizado al menos una compra repetida (es decir, `frequency` mayor que 0). Esto se debe a que el modelo se centra en el gasto de clientes que ya han establecido un patrón de compra.

### Reasoning:
The Gamma-Gamma model is designed for customers with at least two transactions (i.e., `frequency` > 0 in the summary data, where frequency represents repeat purchases). I will filter the `summary_bgnbd` DataFrame to include only these customers before proceeding with the model training.

```python
summary_gg = summary_bgnbd[summary_bgnbd['frequency'] > 0]

print("Primeras filas del DataFrame filtrado para Gamma-Gamma:")
display(summary_gg.head())
```

### Subtask: Entrenar el modelo Gamma–Gamma

Con los datos correctamente filtrados, el siguiente paso es entrenar el modelo Gamma–Gamma. Este modelo aprenderá la distribución del valor monetario de las compras de los clientes.

```python
from lifetimes import GammaGammaFitter

# Ensure monetary_value is strictly positive for Gamma-Gamma model fitting
# This step is crucial because the Gamma-Gamma model assumes positive monetary values.
summary_gg = summary_gg[summary_gg['monetary_value'] > 0]

ggf = GammaGammaFitter(penalizer_coef=0.01)
ggf.fit(summary_gg['frequency'], summary_gg['monetary_value'])

print("Resumen del modelo Gamma-Gamma entrenado:")
print(ggf.summary)
```

## 3. Calcular el CLV Estimado

### Subtask:
Combinar las predicciones del modelo BG/NBD (compras futuras) y del modelo Gamma–Gamma (gasto promedio por compra) para calcular el CLV estimado para cada cliente.

### Reasoning:
To estimate the Customer Lifetime Value, we will multiply the `predicted_purchases` (from BG/NBD) by the `expected_avg_profit` (from Gamma-Gamma). Since `predicted_purchases` are already in `summary_bgnbd` and `expected_avg_profit` is in `summary_gg`, I need to ensure these are correctly aligned, usually by merging or joining the relevant columns based on `CustomerID` before the final multiplication.

```python
summary_gg['expected_avg_profit'] = ggf.conditional_expected_average_profit(
    summary_gg['frequency'],
    summary_gg['monetary_value']
)

# Primero, asegurar que summary_bgnbd y summary_gg estén alineados por CustomerID
# 'predicted_purchases' ya está en summary_bgnbd

# Unir 'expected_avg_profit' de summary_gg al summary_bgnbd
# Usamos un merge left para mantener todos los clientes del summary_bgnbd y añadirles el gasto promedio
clv_estimates = summary_bgnbd.merge(
    summary_gg[['expected_avg_profit']],
    left_index=True,
    right_index=True,
    how='left'
)

# Rellenar los NaN en 'expected_avg_profit' con 0 para clientes que no tuvieron compras repetidas (o valor monetario positivo)
clv_estimates['expected_avg_profit'] = clv_estimates['expected_avg_profit'].fillna(0)

# Calcular el CLV estimado: compras futuras * gasto promedio por compra
clv_estimates['estimated_clv'] = clv_estimates['predicted_purchases'] * clv_estimates['expected_avg_profit']

print("Primeras filas del DataFrame con el CLV estimado:")
display(clv_estimates.head())
```

## 3. Entrenamiento y cálculo del CLV (Usando `lifetimes` `customer_lifetime_value`)

### Subtask:
Calcular el CLV predicho para cada cliente utilizando la función `ggf.customer_lifetime_value`, que combina las predicciones de los modelos BG/NBD y Gamma-Gamma.

### Reasoning:
Instead of manually multiplying `predicted_purchases` and `expected_avg_profit`, the `lifetimes` library provides a more robust method `ggf.customer_lifetime_value`. This function takes the trained BG/NBD model (`bgf`) and the customer's summary statistics to directly estimate CLV over a specified period (`time`) with a `discount_rate`. I will apply this method to the `clv_estimates` DataFrame, ensuring to use the correct columns and parameters.

```python
clv_estimates['CLV_predicho'] = ggf.customer_lifetime_value(
    bgf,
    clv_estimates['frequency'],
    clv_estimates['recency'],
    clv_estimates['T'],
    clv_estimates['monetary_value'],
    time=4,  # meses (como se sugirió en el prompt conceptual)
    discount_rate=0.01 # tasa de descuento
)

print("Primeras filas del DataFrame con el CLV predicho (usando customer_lifetime_value):")
display(clv_estimates.head())
```

## 4. Evaluación y Validación del Modelo CLV

### Subtask: Preparar datos para evaluación

Para evaluar el modelo, necesitamos un DataFrame que contenga tanto los `CLV_real` como los `CLV_predicho` para cada cliente. Esto nos permitirá comparar las estimaciones del modelo con el gasto real de los clientes en la ventana de predicción. Ya hemos realizado este paso en la ejecución anterior, creando el `evaluation_df`.

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Unir clv_dataset (que contiene CLV_real) con clv_estimates (que contiene CLV_predicho)
# clv_estimates tiene CustomerID como índice, así que lo reseteamos para poder hacer el merge.
clv_estimates_reset = clv_estimates.reset_index()

evaluation_df = clv_dataset.merge(clv_estimates_reset[['CustomerID', 'CLV_predicho']], on='CustomerID', how='left')

# Para la evaluación, vamos a considerar solo los clientes para los que el modelo hizo una predicción numérica.
# Los clientes con CLV_predicho NaN son aquellos que no tuvieron frecuencia > 0 en la ventana de observación,
# y por ende, el modelo no predijo un valor de compra.
# Podemos asumir que para estos clientes, si su CLV_real es 0, la predicción de NaN (o 0 después de fillna) es correcta en ese sentido.
# Sin embargo, para las métricas de regresión estándar, es mejor trabajar con valores no nulos.
# Rellenaremos los NaN en CLV_predicho con 0 para que los clientes que no tuvieron actividad o ventas repetidas
# sean incluidos en la evaluación como prediciendo 0, lo cual es razonable si su CLV_real también es 0.

evaluation_df['CLV_predicho'] = evaluation_df['CLV_predicho'].fillna(0)

print("Primeras filas del DataFrame de evaluación:")
display(evaluation_df.head())
```

### Subtask: Visualizar la relación entre CLV Real y Predicho

Crear un gráfico de dispersión para visualizar la correlación entre el `CLV_real` y el `CLV_predicho`. Una línea diagonal (y=x) ayudará a identificar qué tan cerca están las predicciones de los valores reales. Ya hemos realizado este paso en la ejecución anterior.

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='CLV_real', y='CLV_predicho', data=evaluation_df, alpha=0.6)
plt.plot([evaluation_df['CLV_real'].min(), evaluation_df['CLV_real'].max()],
         [evaluation_df['CLV_real'].min(), evaluation_df['CLV_real'].max()],
         color='red', linestyle='--', label='Predicción Perfecta (y=x)')
plt.title('CLV Real vs. CLV Predicho')
plt.xlabel('CLV Real')
plt.ylabel('CLV Predicho')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Subtask: Calcular Métricas de Evaluación

Calcular métricas de regresión comunes como el Error Absoluto Medio (MAE), la Raíz del Error Cuadrático Medio (RMSE) y el Coeficiente de Determinación (R-squared) para cuantificar la precisión del modelo. Ya hemos realizado este paso en la ejecución anterior.

```python
y_true = evaluation_df['CLV_real']
y_pred = evaluation_df['CLV_predicho']

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

print(f"MAE (Error Absoluto Medio): {mae:.2f}")
print(f"RMSE (Raíz del Error Cuadrático Medio): {rmse:.2f}")
print(f"R-squared: {r2:.2f}")
```

### Subtask: Evaluación de Negocio (Top 20% Clientes)

Esta es una métrica clave para el negocio. Consiste en identificar a los clientes que el modelo predice que serán los más valiosos (e.g., el 20% superior por `CLV_predicho`) y luego calcular qué porcentaje del `CLV_real` total es capturado por este segmento. Un buen modelo debería concentrar una alta proporción del `CLV_real` en el top X% de clientes predichos.

```python
# Ordenar el DataFrame por CLV_predicho en orden descendente
evaluation_sorted = evaluation_df.sort_values(by='CLV_predicho', ascending=False)

# Identificar el Top 20% de clientes por CLV predicho
top_20_percent_customers = evaluation_sorted.head(int(len(evaluation_sorted) * 0.2))

# Calcular el CLV Real capturado por el Top 20% de clientes
revenue_captured_by_top_20 = top_20_percent_customers['CLV_real'].sum()

# Calcular el CLV Real total de todos los clientes
total_real_revenue = evaluation_sorted['CLV_real'].sum()

# Calcular el porcentaje de CLV Real capturado
percentage_captured = (revenue_captured_by_top_20 / total_real_revenue) * 100

print(f"CLV Real total: {total_real_revenue:.2f}")
print(f"CLV Real capturado por el Top 20% de clientes predichos: {revenue_captured_by_top_20:.2f}")
print(f"Porcentaje de CLV Real capturado por el Top 20% de clientes predichos: {percentage_captured:.2f}%")
```
